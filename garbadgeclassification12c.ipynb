```python
# ======================================================
# GARBAGE CLASSIFICATION - TRANSFER LEARNING CNN (MobileNetV2)
# MULTI-GPU / BIG BATCH SIZE
# ======================================================

import os
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, Input
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

from sklearn.metrics import classification_report, confusion_matrix

# ======================================================
# DISTRIBUTED STRATEGY (MULTI-GPU)
# ======================================================
strategy = tf.distribute.MirroredStrategy()
print(f"Number of GPUs: {strategy.num_replicas_in_sync}")

# -----------------------------
# CONFIG
# -----------------------------
DATASET_DIR = "/kaggle/input/garbage-classification/garbage_classification"
IMG_SIZE = 224   # MobileNetV2 standard input
BATCH_SIZE = 512  # bigger batch size to utilize VRAM
EPOCHS = 10

classes = [
    "battery","biological","brown-glass","cardboard","clothes",
    "green-glass","metal","paper","plastic","shoes","trash","white-glass"
]

# -----------------------------
# DATA GENERATORS
# -----------------------------
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_gen = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training",
    shuffle=True,
    classes=classes
)

val_gen = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
    shuffle=False,
    classes=classes
)

class_names = list(train_gen.class_indices.keys())
print("Classes:", class_names)

# ======================================================
# BUILD MODEL INSIDE STRATEGY SCOPE
# ======================================================
with strategy.scope():
    base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base_model.trainable = False  # Freeze base model

    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation="relu")(x)
    x = Dropout(0.5)(x)
    output = Dense(len(class_names), activation="softmax")(x)

    model = Model(inputs=base_model.input, outputs=output)
    model.compile(
        optimizer=Adam(learning_rate=0.0001),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )

model.summary()

# -----------------------------
# CALLBACKS
# -----------------------------
callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ModelCheckpoint("garbage_mobilenetv2_dualGPU.keras", save_best_only=True)
]

# ======================================================
# TRAINING
# ======================================================
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks
)

# ======================================================
# ACCURACY & LOSS CURVES
# ======================================================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history["accuracy"], label="Train Acc")
plt.plot(history.history["val_accuracy"], label="Val Acc")
plt.legend()
plt.title("Accuracy Curve")

plt.subplot(1,2,2)
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend()
plt.title("Loss Curve")
plt.show()

# ======================================================
# VALIDATION ACCURACY
# ======================================================
val_loss, val_acc = model.evaluate(val_gen)
print(f"\nValidation Accuracy: {val_acc:.4f}")

# ======================================================
# CLASSIFICATION REPORT
# ======================================================
val_gen.reset()
pred_probs = model.predict(val_gen)
pred_labels = np.argmax(pred_probs, axis=1)
true_labels = val_gen.classes

print("\nClassification Report:\n")
print(classification_report(true_labels, pred_labels, target_names=class_names))

# ======================================================
# CONFUSION MATRIX
# ======================================================
cm = confusion_matrix(true_labels, pred_labels)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# ======================================================
# TEST ON 20 RANDOM IMAGES
# ======================================================
all_images = []
for cls in class_names:
    cls_dir = os.path.join(DATASET_DIR, cls)
    for img_name in os.listdir(cls_dir):
        all_images.append((os.path.join(cls_dir, img_name), cls))

samples = random.sample(all_images, 20)

plt.figure(figsize=(15,10))
for i, (img_path, actual_label) in enumerate(samples):
    img = load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = img_to_array(img)
    img_array = preprocess_input(img_array)
    img_array = np.expand_dims(img_array, axis=0)

    pred = model.predict(img_array)
    pred_label = class_names[np.argmax(pred)]

    plt.subplot(4,5,i+1)
    plt.imshow((img_array[0] + 1)/2)
    plt.axis("off")
    plt.title(f"Actual: {actual_label}\nPred: {pred_label}")

plt.tight_layout()
plt.show()
```
